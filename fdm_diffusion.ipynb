```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
N = 500
L = 100
t = 0.5

# Charge concentration and derivative
c = np.zeros((N, L))
c[0, 45:55] = 1
dc = np.zeros((N, L))
d2c = np.zeros((N, L))

# pn junction
ehc = np.zeros(L)
ehc[45:55] = 1

static = np.ones(L)
static[:10], static[90:] = 0, 0

fig, ax = plt.subplots(1, figsize=(12, 12))

for i in range(N - 1):

    d2c[i] = scipy.ndimage.laplace(c[i])
    # dc[i] = np.linalg.norm(np.gradient(c[i]))
    c[i+1] = (c[i] + t*d2c[i]) * static  # + 0.05*t*dc[i]) * static

ax = sns.heatmap(c, cbar=False)
ax.set(xticks=[], yticks=[], xlabel='x', ylabel='time')
plt.show()
```
